a. Juan debe ser mayor de 35 a√±os, o Pedro debe ser menor de 50 a√±os.

b. La suma de sus edades debe ser un n√∫mero par.

c. Manuel debe tener m√≠nimo 1 nieto.

Ejercicio 4.- N√∫meros mayores a 100.

Escriba una funci√≥n que lea un n√∫mero entero (N) mayor que 100 y valide:

1. Si N es menor de 100 retorne: "Error en el n√∫mero" y vuelva a preguntar la entrada.
2. Si N termina en 4, 7 √≥ 9; diga si es divisible por 3.
3. Si N termina en 1 √≥ 5, divida por 10 y retorne el valor entero,
4. En otro caso divida el n√∫mero entre 3 y retorne su residuo.

```python
#EstablecerLosValores
N=0
while N < 100:
  print('Error en el n√∫mero')
  N=float(input('Escribe un n√∫mero: '))
#Operaciones
ResDiv= N // 10
ResMod=N%3
#CambiarLosValoresDeListaACadenaParaPoderOcuparLosDigitos
cadenaLista=[N]
n=', '.join(str(numero) for numero in cadenaLista)
digFin=n[-3]
#CicloParaEstablecerLasCondiciones
if '4' in digFin or '7' in digFin or '9' in digFin:
  if ResMod==0:
    print('Es divisible entre 3')
elif '1' in digFin or '5' in digFin:
  print('El n√∫mero dividio entre 10 es ',ResDiv)
else:
  print('El residuo es ',ResMod)
```

Ejercicio 5.- MCD y MCM.

El m√°ximo com√∫n divisor.
El m√≠nimo com√∫n m√∫ltiplo.
Almacene los resultados en un diccionario e imprima en pantalla el contenido del diccionario.

Construya una funci√≥n que permita calcular la distancia entre dos pares de coordenadas. Utiliza la librer√≠a math para las funciones trigonom√©tricas y de ra√≠z cuadrada.

```python
#ImportarLaLibrer√≠a
import math
#EstablecerLosValores
LatP1=math.radians(float(input('Pon la latitid del 1¬∞ punto: ')))
LonP1=math.radians(float(input('Pon la longitud del 1¬∞ punto: ')))
LatP2=math.radians(float(input('Pon la latitid del 2¬∞ punto: ')))
LonP2=math.radians(float(input('Pon la longitud del 2¬∞ punto: ')))
#EstablecerLaConstantes&LasOperacionSecundarias
R=6378.1
HavLat=(math.sin((LatP2-LatP1)/2))**2
HavLon=(math.sin((LonP2-LonP1)/2))**2
#OperacionPrincipal
d=2*R*math.asin(math.sqrt(HavLat+math.cos(LatP1)*math.cos(LatP2)*HavLon))
#MostarResultados
print('La distancia es ',d,'km')
```
