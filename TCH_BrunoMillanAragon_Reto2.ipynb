1.- La secuencia del cortador de pizza perezoso ( The Lazy Caterer's sequence).

Se trata un problema clásico de combinatoria y geometría que responde a la siguiente pregunta, ¿cuál es el número máximo de piezas en que se puede dividir una pizza utilizando *n* cortes rectos?
Si tienes una pizza y haces cortes rectos, no necesariamente pasando por el centro y con la posibilidad de cruzarse en cualquier punto, ¿cuál es la mayor cantidad de rebanadas que puedes obtener?. Sabemos que

f(0)=1

f(1)=2

f(2)=4

f(3)=7

Con base en esta información, se puede utilizar la siguiente fórmula de recursión para *n* cortes.

\begin{equation}
f(n)=f(n−1)+n
\end{equation}

Genere un programa en Python que muestre la secuencia de n cortes. Proponga un valor apropiado de cortes y justifique porqué escogió dicho valor.

```python
#utilizamosUnaFuncionParaQue'n'SePuedaModificarConElNúmeroDeCortesQueQueremos
def pizza(n):
#establecerLosValoresDeLasVariables
  x=0
  corte=1
#ciclo'while'ParaQueGenereLosValoresDe'f()'AnterioresA'n'EstoesNecesarioYaQueLosValoresDependenDeLosValoresPasados
  while x <= n:
    corte=corte+x
    x=x+1
#ciclo'if'ParaSóloMostarElValorDe'f(n)'QueQueremosVer
    if x > n:
      print("El número de piezas por hacer",n,"cortes es",corte)
#llamar.A.LaFunciónAdemasDeUn'print'ParaJustificarLa'n'
pizza(5)
print('Con 5 cortes uno puede hacer una figura simétrica pero a partir de 6 la figura ya no es simétrica')
```

2.- Serie de Madhava.

Utilice un ciclo for para estimar el valor de pi ($\pi$) a partir de los primeros 20 términos de la Serie de Madhava, dada por la siguiente expresión:

\begin{equation}
\pi \approx \sqrt{12}\left(
\sum_{k=0}^{19} \frac{(-1)^k}{(2k+1)3^k}
\right)
\end{equation}

```python
#establecerLosValoresDeLasVariables
x=0
pi=0
#crearUnaListaParaAlmacenarLosValoresDeLaSuceción
SerieMad=[]
#Un'n'ParaEstablecerHastaQueTerminaLaSucecion
n=int(input('¿Cuál es el último termino de la serie? '))
#unPrintParaQueSeVeaLosResultadosDeMejorManera
print('La aproximacion de pi por la serie de Madhava es:')
#ciclo'for'ParaAlmacenarLosValoresDeLasuceción
for x in range(n+1):
  Madhava=((-1)**x)/((2*x+1)*(3**x))
  SerieMad.append(Madhava)
#ciclo'for'ParaQueLosValoresDeLaSucecionSeMultipliquenPorLaRaízDe12&ImprimanTodosLosValores
for y in SerieMad:
  pi=pi+y
  piFin=pi*(12**(1/2))
  print(piFin)
```
