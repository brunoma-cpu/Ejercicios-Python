7. Dadas las matrices A y B
Calcule:
* 2B + 3A
* (B + 2A)**2
* A2 + 3B

```python
#instalarLasLibrerias
import numpy as np
#crearLasMatricesConValoresDelEjercicio
MatzA=np.array([[11,7,-1],[-3,5,2],[4,2,-2]])
MatzB=np.array([[1,1,1],[10,3,-5],[-6,-10,-3]])
#crearLaNuevaMatrizDeseadoUtilizandoUnComandoDe'numpy'
Matz71=np.add(np.multiply(2,MatzB),np.multiply(3,MatzA))
Matz72=np.square(np.add(MatzB,np.multiply(2,MatzA)))
Matz73=np.add(np.multiply(2,MatzA),np.multiply(3,MatzB))
#mostarLosResultados
print('2B+3A')
print(Matz71)
print('(B+2A)**2')
print(Matz72)
print('A2+3B')
print(Matz73)
```

8. Genere una función que reciba como entradas las dimensiones de una matriz T[n, m] (3 <= n, m <=5). Asegúrese de validar que sea cuadrada; de lo contrario, debe devolver el mensaje de error La matriz no
es cuadrada y vuelva a preguntar las entradas. Posteriormente reemplace los valores con números aleatorios
e imprima:
* La primera fila de manera inversa.
* La primera columna de manera inversa.
* El primer elemento y el  ́ultimo elemento.
* El valor máximo, mínimo, el promedio de la diagonal y la traza.
* Los valores de T desde T[1,1] hasta T[5,5] (Dependiendo de las dimensiones de la matriz).
* Los valores desde T[1,5] hasta T[5,1] (Dependiendo de las dimensiones de la matriz).
* La suma de los elementos pares recorriendo las filas.
* La suma de los elementos impares recorriendo las columas.
* Imprima los promedios de las filas y columnas.
* La antidiagonal y el promedio de sus valores.

```python
#instalarLasLibrerias
import numpy as np
#definirUnaFunciónQueCreeUnaMatrizAleatoria&QueCumplaConLasCondiciones
def valores():
  #definirA'm'&'n'ConGlobalParaQueSeUtilizenEnOtrasFunciones
  global n
  global m
  n=-1
  m=-2
  #cicloParaQueCumplaConLasCondicines
  while n!=m:
    n=int(input('filas: '))
    #cicloParaCumplirConLaCondiciónde'n'
    while n<3:
      print('filas no validas')
      n=int(input('filas: '))
    m=int(input('columans: '))
    #cicloParaCumplirConLaCondiciónde'n'
    while 5<m:
      print('columas no validas')
      m=int(input('columans: '))
    #condicionalParaCumplirQueSeaCuadrada
    if n!=m:
      print('La matriz no es cuadrada')
  #crearUnaMatrizAleatorea'nxm'ConGlobalParaQueFuncioneEnOtrasFunciones
  global Matz8
  Matz8=np.random.randint(low=1, high=20, size=(n,m))
  #mostarLaMatrizParaVerSusValores
  print(Matz8)
#definirUnaFunciónParaCcumplirConEl1°Incico
def ocho1():
  print("La primera fila de Matz8 de manera inversa:", Matz8[0][::-1])
#definirUnaFunciónParaCcumplirConEl2°Incico
def ocho2():
  print("La primera columna de Matz8 de manera inversa:", Matz8[:, 0][::-1])
#definirUnaFunciónParaCcumplirConEl3°Incico
def ocho3():
  print("Primer elemento:", Matz8[0,0])
  print("Último elemento:", Matz8[-1,-1])
#definirUnaFunciónParaCcumplirConEl4°Incico
def ocho4():
  #crearElResultadoDeseadoUtilizandoUnComandoDe'numpy'
  Max8=np.max(Matz8)
  Min8=np.min(Matz8)
  print("Máximo elemento:", Max8)
  print("Mínimo elemento:", Min8)
  traza8=np.trace(Matz8)
  print("Traza de la matriz:",traza8)
  diago8=np.diag(Matz8)
  promdiago=np.mean(diago8)
  print("Promedio de la diagonal:",promdiago)
#definirUnaFunciónParaCcumplirConEl5°Incico
def ocho5():
  #cicloParaLlamarA.CadaElementoDeLaMatriz
  for x in range(n):
    for y in range(m):
      cinco8=Matz8[x,y]
      print('Elemento [',x,',',y,']:',cinco8)
#definirUnaFunciónParaCcumplirConEl7°Incico
def ocho7():
  #definirLaVariable
  suma7=0
  #cicloParaLlamarA.CadaElementoDeLaMatriz
  for x in range(n):
    for y in range(m):
      cinco7=Matz8[x,y]
      #operacionParaSaberSiEsPar
      modpar=cinco7%2
      #condicionalParaHaceerLaSumaDeLosElementos
      if modpar==0:
        suma7=suma7+cinco7
  #printEnEstáPosiciónParaLlamarAlResultadoSóloCuandoYaTieneATodosLosElementos
  print('Suma de los elementos pares:',suma7)
#definirUnaFunciónParaCcumplirConEl8°Incico
def ocho8():
  #definirLaVariable
  suma8=0
  #cicloParaLlamarA.CadaElementoDeLaMatriz
  for x in range(n):
    for y in range(m):
      cinco8=Matz8[x,y]
      #operacionParaSaberSiEsPar
      modimpar=cinco8%2
      #condicionalParaHaceerLaSumaDeLosElementos
      if modimpar==1:
        suma8=suma8+cinco8
  #printEnEstáPosiciónParaLlamarAlResultadoSóloCuandoYaTieneATodosLosElementos
  print('Suma de los elementos impares:',suma8)
#definirUnaFunciónParaCcumplirConLaPrimeraParteDel9°Incico
def ocho9():
  #definirLaVariable
  suma9=0
  prom9=0
  #cicloParaLlamarA.CadaElementoDeLaMatriz
  for x in range(n):
    for y in range(m):
      cinco9=Matz8[x,y]
      #crearLosValoresParaElPromedio
      suma9=suma9+cinco9
      prom9=suma9/m
    #printEnEstáPosiciónParaLlamarAlResultadoSóloCuandoYaTieneATodosLosElementos
    print('Promedio de los elementos de la fila',x,'es:',prom9)
    #restablecerLosValores
    prom9=0
    suma9=0
#definirUnaFunciónParaCcumplirConLaSegundaParteDel9°Incico
def ocho9a():
  #definirLaVariable
  suma9a=0
  prom9a=0
  #cicloParaLlamarA.CadaElementoDeLaMatriz
  for x in range(n):
    for y in range(m):
      cinco9a=Matz8[y,x]
      #crearLosValoresParaElPromedio
      suma9a=suma9a+cinco9a
      prom9a=suma9a/m
    #printEnEstáPosiciónParaLlamarAlResultadoSóloCuandoYaTieneATodosLosElementos
    print('Promedio de los elementos de la columna',x,'es:',prom9a)
    #restablecerLosValores
    prom9a=0
    suma9a=0
#definirUnaFunciónParaCcumplirConEl10°Incico
def ocho10():
    #crearElResultadoDeseadoUtilizandoUnComandoDe'numpy'
    flip=np.fliplr(Matz8)
    antDiago=np.diag(flip)
    print('La antidiagonal:',antDiago)
#definirUnaFunciónQueLlameA.LasAnteriresFunciones
def mainocho():
  valores()
  ocho1()
  ocho2()
  ocho3()
  ocho4()
  ocho5()
  ocho7()
  ocho8()
  ocho9()
  ocho9a()
  ocho10()
#llamarA.LaFunciónPrincipalConTodosLosIncicios
mainocho()
```
